# 2. _script_

A publicaﾃｧﾃ｣o dos dados do [acervo](https://www.kaggle.com/lexmlacervo/lexml-brasil-acervo) foram realizadas em JSON por ser um formato mais amplamente utilizado nos dias de hoje e, portanto, o escolhemos para coletar todas as urns que precisamos.

```python
# importamos todas as bibliotecas necessﾃ｡rias
import bs4
import requests
from requests.exceptions import ConnectionError
from pathlib import Path
import json
from json.decoder import JSONDecodeError
from time import sleep
from loguru import logger
from tinydb import TinyDB, Query
from typing import Iterable
import re
import gc
```

Definimos a raiz do domﾃｭnio do portal [LexML](https://www.lexml.gov.br/) onde serﾃ｣o feitas as requisiﾃｧﾃｵes e armazenamos todos os arquivos json que serﾃ｣o utilizados para coletar as urns que serﾃ｣o utilizadas nas requisiﾃｧﾃｵes.

```python
urns = Path("./data/federal/")
urns = urns.rglob("*.json")
LOG_FILE = "./log/coleta_metadados_normativos.log"
logger.add(
    LOG_FILE, format="{time:YYYY-MM-DD HH:mm:ss} | {file} {level} | {line} | {message}"
)
```

Em seguida podemos iterar nos arquivos, coletar as urns de cada objeto e assim fazer a chamada da funﾃｧﾃ｣o `get_metadata` que serﾃ｡ responsﾃ｡vel pela coleta dos dados.

```python
urn_retry = {} #container para armazenar uma lﾃｳgica para retry de requisiﾃｧﾃｵes 
for file_ in urns:
    with open(file_, encoding="utf8") as f:
        dados_acervo = json.load(f)
        lista_urns = [u.get("urn") for u in dados_acervo]
        for index, urn in enumerate(lista_urns):
            get_metadata(urn=urn, **{"foldername": file_, "index": index})
```

A funﾃｧﾃ｣o `get_metadata` serﾃ｡ responsﾃ｡vel por realizaﾃｧﾃ｣o uma requisiﾃｧﾃ｣o `get` a url formada pela `BASE_URL` concatenada a `urn` dos normativos, o objeto alvo da nossa coleta ﾃｩ um `json` que encontra-se encapsulado em uma tag `<script>`. Abaixo, apresentamos o cﾃｳdigo da funﾃｧﾃ｣o.

```python
def get_metadata(urn: str, **kwargs):
    BASE_URL = "https://www.lexml.gov.br/urn"
    global urn_retry
    if urn not in urn_retry:
        urn_retry[urn] = 0
    foldername = kwargs.get("foldername")
    index = kwargs.get("index")
    sleep(0.5)
    try:
        urn_retry[urn] += 1
        r = requests.get(f"{BASE_URL}/{urn}")
    except ConnectionError:
        sleep(30)
        if urn_retry[urn] > 5:
            logger.critical(
                f"O programa serﾃ｡ encerrrado por excesso de requests para a mesma urn."
            )
            exit()
        else:
            get_metadata(urn=urn, **{"foldername": file_, "index": index})
    else:
        if r.status_code == 200:
            soup = bs4.BeautifulSoup(r.text, features="html.parser")
            scripts = soup.find_all("script")
            tag_script = [
                str(s) for s in scripts if re.search("application\/ld\+json", str(s))
            ]
            if tag_script:
                tag_script = tag_script[0]
                first_slice = tag_script.find("[")
                last_slice = tag_script.find("</script>")
                try:
                    metadata = json.loads(tag_script[first_slice:last_slice])
                except JSONDecodeError:
                    try:
                        first_slice = tag_script.find("{")
                        last_slice = tag_script.find("</script>")
                        metadata = json.loads(
                            "[" + tag_script[first_slice:last_slice] + "]"
                        )
                    except JSONDecodeError:
                        logger.error(f"Erro na conversﾃ｣o em JSON da urn {urn}.")
                        pass
                    else:
                        save_file(
                            data=metadata,
                            filename=index,
                            foldername=foldername,
                            **{"urn": urn},
                        )
                else:
                    save_file(
                        data=metadata,
                        filename=index,
                        foldername=foldername,
                        **{"urn": urn},
                    )

            else:
                logger.info(f"A urn {urn} nﾃ｣o possui metadados.")
        else:
            logger.error(f"Requisiﾃｧﾃ｣o para urn {urn} com status code {r.status_code}.")
```

Por ﾃｺltimo, mas nﾃ｣o menos importante, implementamos a funﾃｧﾃ｣o `save_file` que serﾃ｡ responsﾃ｡vel por persistir em disco nossos dados.

```python
def save_file(data: dict, filename: str, foldername: str, **kwargs):
    urn = kwargs.get("urn")
    get_year_folder = str(foldername).split("_")[-1].split(".")[0]
    BASE_PATH = Path(f"./data/metadados/{get_year_folder}")
    BASE_PATH.mkdir(parents=True, exist_ok=True)
    BASE_PATH = BASE_PATH / f"{get_year_folder}_{filename}.json"
    try:
        with open(BASE_PATH, mode="w", encoding="utf8") as _:
            json.dump(data, _, ensure_ascii=False)
    except ValueError:
        logger.error(f"Erro no dump para urn {urn}.")
    else:
        logger.success(f"Metadados da urn {urn} salvo com sucesso.")
```

Agora, podemos agregar os dados em uma soluﾃｧﾃ｣o de banco de dados leve orientada a documentos como o [`Tinydb`](https://github.com/msiemens/tinydb) de forma que podemos consultar os dados de maneira estruturada. Primeramente, vamos estruturar uma funﾃｧﾃ｣o que irﾃ｡ gravar os dados no banco.

```python
def insert_into_db(list_of_files: Iterable, db: TinyDB, key: str) -> None:
    container = []
    for f in list_of_files:
        if len(container) > 500000:
            db.insert_multiple(container)
            container = []
            gc.collect()
        with open(f, "r", encoding="utf8") as _:
            data = json.load(_)
            for d in data:
                if key == "metadados":
                    try:
                        d["keywords"] = d["keywords"].split(",")
                    except KeyError:
                        print(
                            f"Nﾃ｣o foi possﾃｭvel encontrar o campo keywords no arquivo {f}."
                        )

                container.append(d)

    db.insert_multiple(container)
```

Por fim, podemos agora listar os arquivos a serem inseridos no banco e fazer uma chamada a funﾃｧﾃ｣o recﾃｩm criada. Alﾃｩm disso, os metadados do acervo do portal LexML haviam sido publicados no [kaggle](https://www.kaggle.com/lexmlacervo/lexml-brasil-acervo) de forma desagregada, isto ﾃｩ, foi disponibilizados vﾃ｡rios arquivos `json`. Portanto, vamos organizﾃ｡-los em um ﾃｺnico arquivo de banco de dados para facilitar o seu uso e consulta.

```python
db_metadados = TinyDB("./data/db/metadados.json") #instﾃ｢ncia do tinydb para os metadados de normas federais
files_metadados = Path("./data/metadados/")
files_metadados = files_metadados.rglob("*.json") #arquivos json dos metadados das normas federais

db_acervo = TinyDB("./data/db/acervo.json") #instﾃ｢ncia do tinydb para os dados do acervo LexML.
files_acervo = Path("./data/json/")
files_acervo = files_acervo.rglob("*.json") #arquivos json do acervo LexML

params = {
    "metadados": (db_metadados, files_metadados),
    "acervo": (db_acervo, files_acervo)
}

for key, data in params.items():
    insert_into_db(list_of_files=data[1], db=data[0], key=key)
```

# 4. aplicaﾃｧﾃ｣o em aprendizado de mﾃ｡quina

Ao meu ver o primeiro uso desse dataset para fins de `machine learning` ﾃｩ um problema de [classificaﾃｧﾃ｣o multi-label](https://en.wikipedia.org/wiki/Multi-label_classification).  Para cada normativo, temos um campo `keywords` que recebe mﾃｺltiplas tags que estﾃ｣o relacionadas ao assunto dessa norma jurﾃｭdica. Alﾃｩm disso, temos o campo `about.description` que registra a referida **ementa**.

Sabido que esse processo de _tagging_ ﾃｩ, geralmente, realizado por meio de anotaﾃｧﾃ｣o manual, isto ﾃｩ, um profissional do meio jurﾃｭdico ou alguﾃｩm atue com publicaﾃｧﾃ｣o de normativos legais irﾃ｡ decidir quais `tags` sﾃ｣o mais apropriadas a esse documento legal. Essa sistemﾃ｡tica de trabalho, possui algumas desvantagens, podendo-se citar algumas, dentre as quais:

- Sujeito a falhas, como a maior parte dos procedimentos manuais;
- Falta de uniformidade na classificaﾃｧﾃ｣o;
- Necessidade de uma equipe multidisciplinar disponﾃｭvel para anotaﾃｧﾃ｣o;
- Para o resultado da classificaﾃｧﾃ｣o ser acurado e satisfatﾃｳrio, exige-se um alto custo de homem-hora;

As caracterﾃｭsticas elencadas acima, jﾃ｡ mostram razﾃ｣o suficiente para que a tecnologia, por meio de um modelo de aprendizado de mﾃ｡quina auxilie os profissionais no desempenho de suas funﾃｧﾃｵes.

> O papel da tecnologia no mundo jurﾃｭdico foi recentemente abordada no podcast, [hipsters ponto tech](https://hipsters.tech/tecnologia-no-juridico-hipsters-208) 太.

Assim, com o `dataset` apresentado acreditamos ter um potencial para o desenvolvimento de um _[data product](https://www.oreilly.com/radar/drivetrain-approach-data-products/)_ que poderia ser acoplado a um sistema prﾃｩ-existente de publicaﾃｧﾃ｣o de atos normativos. Assim, o usuﾃ｡rio abdicaria da responsabilidade de inserir manualmente as `tags` mais apropriadas, e o sistema faria a consulta a API do modelo de aprendizado de mﾃ｡quina e o seu front-end jﾃ｡ apresentaria a sugestﾃ｣o das `tags` mais apropriadas para a norma em anﾃ｡lise. Assim, a responsabilidade do usuﾃ｡rio seria apenas de *validar* a sugestﾃ｣o do modelo, de modo a desonerar parte do trabalho intelectual do profissional.

Portanto, a proposiﾃｧﾃ｣o ﾃｩ a criaﾃｧﾃ｣o de um modelo de classificaﾃｧﾃ｣o multi-label, treinando-o sobre as ementas de cada norma. Nas prﾃｳximas postagens, esperamos continuar explorando esses dados e iniciar o processo de treinamento do modelo. Atﾃｩ mais! 
